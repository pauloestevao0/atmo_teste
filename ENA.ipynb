```python
from pandas.core.arrays.datetimelike import timedelta
import pandas as pd
import numpy as np
import datetime
import logging

#importar csv tlv_vazao
url_vazao = "https://raw.githubusercontent.com/pauloestevao0/atmo_teste/main/tbl_vazao.csv" 
vazao = pd.read_csv(url_vazao, sep=",") 

#importar csv tlv_postos
url_postos = "https://raw.githubusercontent.com/pauloestevao0/atmo_teste/main/tbl_postos.csv"
postos = pd.read_csv(url_postos, sep=",")

#juntando arquivos
joined_files = pd.merge(vazao, postos, on='num_posto', how='left')
lista_df = pd.DataFrame(joined_files)

#alterando formato data
lista_df.loc[:,'dat_medicao'] = pd.to_datetime(lista_df['dat_medicao'])


#calculando ENA
lista_df['ENA'] = lista_df['val_vaz_natr'] * lista_df['val_produt']

#filtrando bin_ena > 0
filtro = lista_df['bin_ena']>0
filtered_list = lista_df[filtro]


#gerando pivot com mdia diria
diaria = pd.pivot_table(filtered_list, index=['dat_medicao'], values = ['ENA'], columns = ['num_ssis', 'nom_ssis'], aggfunc = np.average)

#agrupando por mes
grouped_list = filtered_list.groupby(['num_ssis','nom_ssis',pd.Grouper(key = 'dat_medicao', freq = 'MS')]).mean()

#excluir colunas irrelevantes
mensal = grouped_list.drop(columns = ['num_posto','val_vaz_natr','bin_ena','val_produt'])

start = datetime.datetime.now()


print("Inicializando o processo de captura de dados - {}".format(start))
print("Processo finalizado com sucesso - {}".format(datetime.datetime.now()))
display(mensal)
print(diaria.to_markdown())


```
