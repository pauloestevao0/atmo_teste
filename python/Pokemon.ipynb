```python
import requests 
import pandas as pd 
import datetime
import logging

#buscando geraçao
def get_generation(gen):
  api = "https://pokeapi.co/api/v2/generation/{}".format(gen)
  response = requests.get(api)
  listapoke = response.json()

  names = []
  for i in listapoke['pokemon_species']:
    names.append(i['name'])

  df = pd.DataFrame(names, columns = ['pokemon_name'])

  return(df)

#buscando nomes pokemons
def get_pokemons(names):
  df = pd.DataFrame()
  for i in names['pokemon_name']:
    df_temp = pd.DataFrame()
    api = "https://pokeapi.co/api/v2/pokemon/{}".format(i)
    response = requests.get(api)
    listapoke = response.json()
    abilities = []

    df_temp.loc[0,'name'] = i

    #Pegando dados de habilidade do pokemon
    for i in listapoke['abilities']:
      abilities.append(i['ability']['name'])
    str_abilities = ','.join(abilities)
    df_temp.loc[0,'abilities'] = str_abilities
   
    #Pegando dados do tipo do pokemon
    types = []
    for i in listapoke['types']:
      types.append(i['type']['name'])
    str_types = ','.join(types)
    df_temp.loc[0,'types'] = str_types

    #Pegando locais onde pode ser encontrado 
    location_encounter = requests.get(listapoke['location_area_encounters']).json()
    locals = []
    for i in location_encounter:
      locals.append(i['location_area']['name'])
    str_locals = ','.join(locals)
    df_temp.loc[0,'locals'] = str_locals




    df = df.append(df_temp)
  return df

def main():
  now = datetime.datetime.now()
  print("Inicializando o processo de captura de dados - {}".format(now))
  names = get_generation('generation-i')
  pokemons = get_pokemons(names)
  print("Processo finalizado com sucesso - {}".format(datetime.datetime.now()))
  return pokemons

main()
```
